```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("zoo.csv")
data.head()
```

```python
classes = ['Mammal', 'Bird', 'Reptile', 'Fish', 'Amphibian', 'Bug', 'Invertebrate']
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop('class_type', axis=1).drop('animal_name', axis=1), 
    data['class_type'], 
    test_size=0.33, random_state=42
)
```

```python
X_train.head()
```

```python
len(X_train)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)

tree.plot_tree(model)
```

```python
import matplotlib.pyplot as plt


model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)


fig=plt.figure(figsize=(25,20))
tree.plot_tree(model,feature_names=X_train.columns,class_names=classes,filled=True)
```

```python
pred = model.predict(X_test)
pred[:10]
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred, target_names=classes))
```

```python
print("Actual\t\t\t\tPrediction")
y_test = list(y_test)

for i in range(len(y_test[:25])):
    print(f"{classes[y_test[i]-1]}\t\t\t\t{classes[pred[i]-1]}")
```
